методы:

    1. asyncio.get_event_loop() - возвращает работающий ивент луп или создает его, если речь идет о мэйн треде и если set_event_loop() не был вызван, но ГЕТ сам может вызвать СЕТ, если надо, и возвращаемый объект можно записать в некоторую переменнцю, на которой можем вызвать 3 пункт.
    2. asyncio.get_running_loop() - возвращает работающий ивент луп, но никак не создает его, если вызова этого метода ивент луп не был создан, то будет ошибка рантаймерор
    3. loop.run_until_complete(fn()) - метод, который вызывается на объекте ивент луп, loop здесь - это некоторая переменная
    так вот если у нас есть переменная, которуя мы получили,вызвав гет ивент луп, мы можем вызвать на нем ран антил комплит, запустив какой-нибудь корутин. корутин в этом контрукторе находится под именем fn(), соответственно этот loop будет работать до завершения fn(), а в нем может быть спрятана всякая тема, целую прогу можно запихнуть, просто как точка входа
    3.1 если нам в корутине нужно получить объект ивент лупа, то лучше вызывать гет ранинг ивет луп
    4. loop.close() - "закрывает" иветлуп, но тут стоит иметь в виду, что прежде чем закрыть, все корутины должны уже завершить свою работы. Метод очищает все очереди, все закрывает, больше ничего не исполняется. На лупе уже ничего нельзя будет вызвать
    5. asyncio.run(fn()) - делает все за нас, так сказать, включает в себя и гет ивент, и клоуз, и кучу всего, при этом даже умнее, чем мануальное управление. Заботится обо всех возможных заботах и проблемах лучше,чем мы сами
    6. loop.create_task(fn()) - позволяет поставить корутин в очередь, но не запускать его
    7. loop.run_forever() - запускает ивентлуп пока не будет вызван метод stop()
    8. loop.stop() - останавливает луп, после него луп опять можно запустить, является обратимой
    9. loop.is_closed() / loop.is_running() - методы инстанции, которые показывают, закрыт ли метод и в работе ли он находится
