future определяет метод cancel()
обычно мы работаем с задачами, а не фьючерами, 

если задача еще не была запущена, а только добавлена в очередь на запуск,и тут мы вызывает cancel() на соответствующем фьючере, который соотносится с определенным таском(задачей), то ее запуск будет отменен, 
cancel() пробрасывает в отменяемой задаче CanceledError, 

задача может проверять CanceledError,
кооперативная отмена есть

future.cancel() работает только на задачи в статусе пендинг, если применить на финишд, то вернет фолс

некооперативная отмена базируется на исключениях, хотя кооперативная тоже может базироваться на них


future = asyncio.gather(*aws, return_exceptions = False) - фолс стоит по умолчанию, 
*aws - это список тасков, можно передать и список корутинов
таски возвращают фьючер(?)
1. исключение пролетает до await, остальные таски не отменяются
2. если стоит True, то таски не отменяются, исключения идут в результаты, а не выскакивают ошибкой
3. явный cancel() на pending-тасках отменяет их, 
4. делать кэнсел на фьючер опасно, вообще мы можем передать в gather список корутинов, а можем корутины завернуть в таски и передать список тасков, и вот если мы передаем не список корутинов,а список тасков, и соответственно мы имеем к ним доступ вне gather ,и при определенных обстоятельствах мы можем вызывать на самих тасках cancel(), не на фьючере,который возвращает gather, а на самих тасках, которые мы передали в gather, в итоге делать так опасно, потому что это дело совсем неопределенное. Поэтому в gather лучше передавать список тасков и делать cancel() на них, а не на фьючере, который возвращается методом gather, 

wait тоже взвращается фьючер, получает список тасков, таймаут, 

asyncio.wait(aws, timeout, return_when = ALL_COMPLETED)

алл_комплитед стоит по умолчанию, и в этом случае  wait вернет значение только когда все фьючеры завершатся или будут отменены

можно передать FIRST_COMPLETED, когда первая завершится или отменится

можно передать FIRST_EXCEPTION, когда ожидается первое исключение. если исключения не было, поведение будет аналогично ол_компл

wait не отменяет таски по истечению таймаута
