1. ошибки можно ловить внутри потока
2. можно ловить в вызыващем потоке, в отличие от трединга это сделать проще, и через фьючер, и через эуэйт
3. если исключение не пронблюдали, то есть результат не был взят, то оно просто проглатывается, та корутина, которая выбросила исключение, прекращает исполнение, но приложение также не убивается
4. await пробрасывает исключение, то есть если из корутины выскочило исключение,и мы сделали await на этой корутине, то await пробросит исключение, и приложение умрет
5. future.result() тоже пробрасывает исключение, если резалт взяли в мейнтреде, то приложение тоже умирает
6. async for тоже пробрасывает исключение прям сразу, внутрь цикла мы не попадем, то есть надо оборачивать сам async for в трай эксепт, чтобы это исключение отловить
7. колбэк - это некая ф-я, которая будет вызвана после завершение таска, как фьючер перейдет в состояние ДАН, навешивается он с помощью add_done_callback(), если же мы в колбэке запрашиваем результат, и он выбрасывает исключение, а мы его не обрабатываем, то приложение не убивается
8. gather(return_exceptions=True) позволяет взять результаты из успешно завершенных задач иполучить список исключений и проваленных,
