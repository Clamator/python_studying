тут у нас есть терминал, который обращается к библиотеке и получает результат
смысл в том, что в asyncio.run(main()) мы запускает ивентпул, 

этим вызовом мы запускаем мэйн, который создает объект класса и вызывает на нем метод старт который записывает в переменную "результат" результат

когда вызывается метод старт, в нем появляются две переменные - ЛУП и Фьюче
луп находит или запускает ивентлуп, в это время во фьюче мы проставляем заглушку под результат

после этого запускается поток, который в качестве аргумента принимает функцию, и еще два аргумента для этой самой ф-и, после этого запускается старт, но не метод класса, а метод для ивентлупа

в это время, когда в потоке вызывается ф-я, в самой этой ф-и принимаются два только что созданных Луп и Фьюче, 
на Луп вызывается кол_сун_тс, передается фьюче и методом "сет_резалт" и прописывается тот резк, который мы хотим указать

все

```python
import asyncio
import threading
import time


class Terminal:

    async def start(self):
        #  тут мы сами будем создавать заглушку фьюче
        #  тут мы породили фьюче, из любой синхронной ф-и можно вызвать асинхронную, создав ивентлуп
        loop = asyncio.get_event_loop()  # можно гет_ранинг_луп(), здесь мы нашли ивентлуп
        future = loop.create_future()  # создали заглушку фьюче,в который будет записываться результат
        t = threading.Thread(target=self.run_cmd, args=(loop, future,))  # делается все тут, арги переданы
        t.start()
        return await future

    # этот метод будет запущен  в отдельном потоке и должен будет установить результат вызова метода старт
    def run_cmd(self, loop, future):
        # здесь делается колбэк - результат устанавливается не в том потоке, в котором порожден луп и фьюче
        time.sleep(1)
        loop.call_soon_threadsafe(future.set_result, 1)  # именно это и устанавливает результат


async def main():
    t = Terminal()
    result = await t.start()  # метод класса
    print(result)


if __name__ == '__main__':
    asyncio.run(main())  # значит, тут мы создаем ивентлуп?
```
