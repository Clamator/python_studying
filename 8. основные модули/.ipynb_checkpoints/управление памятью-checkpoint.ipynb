нужно понимать, как яп обращается с памятью и объектами, чтобы не было вопросов, почему прога потребляет много памяти, и почему потребление со временем растет

!опер.память разделяется на стек и кучу
стек удобен для управления локал переменными, которые входят в область, работает по принципу последний зашел - первый ушел

!когда мы объявляем переменные, мы связываем имена с объектами в куче
имена обычно лежат в стеке, а объекты в куче

! ссылка - это имя, указывающее на объект в куче
объекты лежат в куче
имена - это метки объектов, и один объект мб помечен множеством меток

допустим, у нас есть имя Х, кот лежит в стеке, и объект 100, кот находится в куче
если они связаны, то к ним можно применить такое понятие как референс каунт - подсчет количества ссылок
р.к равен 1,
если мы создаем имя Y и связываем с 100, то новая 100 не создается, Х и У вдвоем связаны с одним сто, при этом р.к обновляется до двух

если сы создадим далее список Z = [100,100], то он также будет ссылаться на первую сотню, и тогда р.к. будет равен 4
если мы удаляем Z, то ссылки удалятся, р.к. = 2
при переприсвоении новых значений ссылки прекращают существование, 

если на объект никто не ссылается, он удаляется

есть две стратегии сбора мусора:
    р.к. 
    трейсинг
    
проблема р.к. - подсчет ссылок не работает в циклических ссылках, если а = б, а б = а, то счетчик р.к. никогда не становится меньше единицы, поэтому гарбедж коллектор их никогда не вычистит, поэтому здесь и нужен трейсинг

трейсинг - строит графы (лес, дерево объектов) объектов, отслеживает их поколения
трейсинг строит эти деревья т.н. и отслеживает взаимосвязь м/у объектами,если он видит циклические ссылки, но никакие корневые объекты не ссылаются на них, то такие объекты будут удалены

иногда запускается гар. кол., который маркирует и удаляет осиротевшие объекты, также он группирует объекты по трем поколениям
новые 0
средние 1
старые 2
когда достигается порог потребления памяти, запускается Г. К.

когда объект переживает сборку мусора, он переходит в более старшее поколение

если растет потребление памяти, надо проверять на наличие цикл. ссылки, да и вообще стараться их избегать
