ужасы многопоточки: подробнее в следующий видео
    есть race - гонка
    и есть deadlock - взаимная блокировка

гонка - это когда у нескольких потоков есть доступ к определенной переменной или чему бы то ни было, так вот, если потоки работают асинхронно, а мб и не только в этом случае, может случиться та самая гонка, в чем ее суть.

допустим, один поток добавляет к переменной что-то, второй - умножает результат на два
из-за этой гонки, т.к. мы ее не контролируем, может сначала первый, а потом второй поток отработать, и в итоге получатся разные результаты

есть такая штука как примитив синхронизации. их много,и они нужны для разных случаев

1. основной, мб самый частоиспользуемый примитив - Lock - он дает эксклюзивный доступ к ресурсу
то есть пока один поток работает с этим ресурсом, никакой другой поток не имеет доступа, ни на чтение, ни на запись
в этом лучае потоки друг другу не мешают 

2. RLock - то же самое, что и лок, но он позволяет потоку входить в самого себя,
позволяет reentrance поток, взявшему лок
reentrance - возвращение

3. Event - позволяет в одном потоке ожидать сигнал от другого потока

4. Condition объединяет Event и Lock

5. Semaphore - ограничивает к ресурсу количество потоков

6. BoundedSemaphore - тут немного сложнее, надо разобраться с п. 5 сначала

7. Barrier - ожидает завершения n потоков

примитивы синхронизации находятся в модуле threading
