атомарные операции - это операции, который выполняются целиком. Когда атомарная операция происходит, мб в ней меняются некоторые данные, другой поток в нее вмешаться не может, и не может сам повлиять на те данные, которые модифицируются
интерпретатор переключает операции только между инструкциями байт кода, сами инструкции неделимы

важно ли знать, что существуют атомарные операции? наверное, лучше знать, однако не стоит опираться на них ппри проектировании многопоточного кода
____
явные локи будут явно документировать и выжарать то, что происходит в коде, будет понятно, что тут делится какой-то ресурс

опытный программист может использовать лок-фри подход, ну то есть надо быть прям спецом. Для этого должны быть веские причины, так же понятно, что локи замедляют исполнение кода

лок-фри в целом является боле производительным подходом
____
есть ряд причин, почему не желательно применять лок фри.
одна из них - придет другой прогер и может не разобраться,
также возможны всякие изменения в интерпретаторе, что он атомарную уже не будет воспринимать как атомарную
___
использовать лок фри имеет смысл при безвыходности и для оптимизации производительности, если уже все сделано, оптимизированно как можно, но этого недостаточно. 
но на начальных этапах развития беспокоитьс об этом не надо
